```python
import pandas as pd
import numpy as np
```

```python
customers= pd.read_csv("olist_customers_dataset.csv")
geolocation=pd.read_csv("olist_geolocation_dataset.csv")
items=pd.read_csv("olist_order_items_dataset.csv")
payments=pd.read_csv("olist_order_payments_dataset.csv")
reviews=pd.read_csv("olist_order_reviews_dataset.csv")
reviews=pd.read_csv("olist_order_reviews_dataset.csv")
orders=pd.read_csv("olist_orders_dataset.csv")
products=pd.read_csv("olist_products_dataset.csv")
sellers=pd.read_csv("olist_sellers_dataset.csv")
category=pd.read_csv("product_category_name_translation.csv")
```

```python
df_1= reviews.merge(orders,left_on = "order_id", right_on = "order_id",
                          how="inner")
df_2= df_1.merge(payments,left_on = "order_id", right_on = "order_id",
                          how="inner")
df_3= df_2.merge(customers,left_on = "customer_id", right_on = "customer_id",
                          how="inner")
df_4= df_3.merge(items,left_on = "order_id", right_on = "order_id",
                          how="inner")
df_5= df_4.merge(products,left_on = "product_id", right_on = "product_id",
                          how="inner")
df_6= df_5.merge(sellers,left_on = "seller_id", right_on = "seller_id",
                          how="inner")
df_6.head()
```
